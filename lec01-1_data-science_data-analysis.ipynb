### 데이터

#### 데이터의 정의
- 데이터란 질적인 혹은 양적인 변수들의 집합
- 의사결정에 적합한 정보를 추출하기 위해서 사용하는 원천

#### 정보
- 관찰이나 측정을 통해 수집한 자료를 당면한 문제 해결을 위해 체계적으로 정리한 것
- 데이터와 비교하여 상대적으로 넓은 의미
- 의사결정을 위한 소통의 근거가 됨

#### 지식
- 데이터나 정보와 비교해서 경험의 측면이 더욱 강조된 개념
- 경험을 축적하는 데에만 그치지 않고 획득된 정보의 관계성까지 고려
- 특정한 주제에 관해 정보를 다루면서 발생하는 여러 가지 이벤트들을 접하게 되는데, 이 과정에서 수많은 축적된 경험이나 파악된 관계성으로부터 도출되는 것이 지식
  - 서울에서 부산까지 최단 거리를 측정  >  이는 데이터
  - 서울에서 부산까지 도달하는 가장 좋은 방법  >  지식의 범주에 해당
  
#### 데이터와 정보, 지식
- 데이터는 정보나 지식을 뒷받침하는 재료
  - 올바른 정보나 지식을 축적하기 위해서는 신뢰도 높은 데이터가 바탕
- 올바른 데이터의 수집 및 해석, 가공 및 정제 등의 절차가 매우 중요


### 데이터의 유형

#### 정형 데이터와 비정형 데이터
##### 정형 데이터
- 데이터의 속성이나 구조가 상대적으로 명확하고 체계화된 데이터
  - 데이터의 속성을 파악하는 데 제약이 적으며 데이터 수집이나 가공에 큰 어려움이 없고
  - 기존 데이터 분석도구로 분석이 용이한 데이터
- 데이터베이스 테이블과 같이 미리 정해진 형식과 구조에 따라 저장된 데이터

  
##### 비정형 데이터
- 형태가 매우 다양하며 데이터의 수집 과정이나 가공 정제 과정에 많은 노력이 필요한 데이터
  - 전통적인 분석도구를 이용하여 분석하기에 많은 제약이 따르므로 별도의 도구를 개발하여 분석하는 경우가 많음
  - 경우에 따라 비정형 데이터를 가공 또는 형태 변환 등의 기법을 사용하여 정형 데이터 형태에 가깝게 변환한 후 정형 데이터 분석 방법 적용
- 텍스트 데이터, 이미지 데이터, 영상 데이터, 음향 데이터, 이러한 형태가 융합된 데이터


#### 정형 데이터의 구분
##### 범주형 데이터 (Categorical data): 수치로 측정하고 표시하는 것이 불가능한 데이터. 질적 속성을 기반으로 범주로 구분 가능한 데이터
- 명목형 데이터 (Nominal data): 순서가 없이 어떤 특성을 지니고 있는 관측치로 이루어진 데이터
  - 우편번호, 성별, 인종, 혈액형, 교과목 코드, 질병 코드
  - 순서를 매길 수 없음
  - 셀 수 있음 

- 순서형 데이터 (Ordinal data): 수치적 정밀도는 제한적이지만 크고 작은 수량적 크기나 의미적인 순서 비교가 가능한 데이터
  - 학력 데이터 (중학교 졸 이하, 고등학교 졸, 대학교 졸, 대학원 졸 등의 구분)
  - 리커트(Likert) 척도 데이터 (설문조사에서 많이 사용하는 5점, 7점 척도 설문 응답 데이터)
  - 인접한 범주 간의 크기 비교가 정확한 간격으로 이루어질 수 없다는 제약이 있음

##### 수치형 데이터 (Numerical data): 수치적 크기를 측정한 관측치로 이루어진 데이터
- 이산형 데이터 (Discrete data): 데이터와 데이터 사이가 연속적이지 않고 끊어져 있음. 정수의 형태로 표현 가능
  - 주차장의 주차 가능 대수, 가족 구성원 수, 신입생 수

- 연속형 데이터 (Continuous data): 데이터와 데이터 사이에 끊어짐이 없이 연속적으로 이어져 있음. 실수로 표현 가능
  - 길이, 무게, 부피, 온도

### 데이터의 형태

#### 벡터: 1차원 데이터
- 한 가지 속성(변수)에 대한 값을 나타내는 한 개 이상의 원소로 구성된 데이터
- 예: 온도 데이터

#### 행렬: 2차원 데이터
- 여러 가지 속성(변수)에 대한 값을 나타내는 데이터
- 예: 학생들의 국어, 영어, 수학, 과학 시험 성적

#### 데이터 프레임
- 행렬과 유사한 2차원 데이터이나 각 속성(변수)가 다른 자료형의 값을 가질 수 있음
- 예: 학생들의 이름(문자형), 나이(숫자형), 학번(숫자형), 성별(문자형)

#### 다차원 데이터
- 3차원 이상의 데이터

### 데이터 과학 (Data Science)
- 데이터를 과학의 대상으로 다루는 다양한 접근법을 의미
  - 체계적으로 데이터를 축적하는 것
  - 과학적인 분석의 틀을 설계하는 것
  - 분석 결과를 정확하게 표현하고 해석하는 것 등
 


- 정형, 비정형 형태를 포함한 다양한 데이터로부터 지식과 인사이트를 추출하는 과정에서 과학적 방법론, 프로세스, 알고리즘, 시스템을 동원하는 융합 분야
  - 데이터를 통해 실제 현상을 이해하고 분석하며 통계학, 데이터 분석, 기계학습과 연관된 방법론을 통합하는 개념으로 정의되기도 함
  - [출처: 위키피디아](https://ko.wikipedia.org/wiki/%EB%8D%B0%EC%9D%B4%ED%84%B0_%EC%82%AC%EC%9D%B4%EC%96%B8%EC%8A%A4)
  
- 데이터를 수집/분석/처리하여 유의미한 정보를 추출하고 활용하는 과학적 방법론, 프로세스, 시스템 등을 포함하는 학제 간 연구 분야
  - 정형, 비정형을 포함한 다양한 유형의 데이터를 분석하여 현상을 이해하는 체계적인 이론과 기술들을 통칭함
  - [출처: 정보통신용어사전](http://terms.tta.or.kr/dictionary/dictionaryView.do?word_seq=178860-2)

#### 과학적 접근법/방법론
- 가설: 과학적 접근법의 시작
  - 가설: 검증의 대상이 되는 질문
  - 마케팅에서 어떤 형태의 사람들이 어떤 상품을 구매할 성향이 강한지 여부 판단
  - 건강과 관련하여 커피를 어떻게 마시는 것이 심혈관 예방에 효과가 있는지 비교
- 가설의 검증
  - 가설은 참이거나 거짓일 가능성이 있음
  - 관련된 데이터를 체계적으로 수집하여 증거를 축적한 뒤 검증
- 이론의 정립
  - 데이터의 수집과 가설 검증의 과정과 같은 구체적 현상에 대한 경험 과정을 '관찰'이라고 함
  - 여러 차례의 관찰, 즉 데이터가 쌓이면 '경험의 일반화'에 이르게 되며 이는 '이론'으로 정립될 수 있음

#### 과학적 접근 방법과 데이터 과학

- 데이터 과학은 연구 질문인 가설에 답하기 위한 경험적 접근 방식과 관련
  - 즉, 관찰, 측정, 데이터 수집 및 해석 등의 단계와 경험적 일반화에 이르는 과정에 초점을 맞춘 과학적 접근 방법
- '데이터가 스스로 말하게 하는 것'에 중점을 둠
- 데이터 과학에서는 방법론적 관점에서 경험적 조사를 원활하게 수행하기 위한 방법, 도구 및 접근 방식을 개발하는 것이 중요
- 주요 목표: 데이터 수집 및 저장 방법 연구, 데이터 분석 방법의 이론적 정립, 장점 및 단점의 이해를 통한 평가, 개선 방안에 대한 연구

#### 데이터 과학자
- 1) 데이터의 수집과 저장에 관한 이해를 바탕으로
- 2) 이를 가공하고 처리하는 기술적인 지식을 갖추고 분석 과정에서 주도적인 역할을 하면서
- 3) 과학적 의사결정에 이르게 하는 전문가


- 등장 배경
  - 빅데이터 시대의 도래와 함께 데이터 분석이라는 한정된 영역을 뛰어넘는 탁월한 지식과 능력을 지닌 전문가의 필요성 증대
  - 데이터에 대한 이해력, 분석 능력뿐만 아니라 분석 결과의 해설 및 통찰력까지 지닌 전문가의 출현이 요구됨

- 데이터 과학자의 필요 능력

### 데이터 과학과 데이터 분석

#### 데이터 분석
- 유용한 정보를 발견하고 결론을 유추하거나, 의사 결정을 지원하는 것을 목표로 데이터를 조사, 정제, 변환, 모델링하는 과정
  - Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making. (https://en.wikipedia.org/wiki/Data_analysis)
  
- 데이터를 해석하여 현재의 상태를 이해하고 이를 바탕으로 유용한 결론이나 통찰을 도출하는 것
- 주로 과거 데이터에 초점을 맞추고 데이터를 통해 패턴을 찾거나 요약하는 작업에 중점
- 데이터 분석은 데이터 과학보다 좁은 범위의 개념으로 봄


#### 데이터 과학
- 데이터를 통해 복잡한 문제 해결을 위해 최선의 솔루션을 만드는 것
- 수집된 데이터를 분석하고 그 결과를 바탕으로 예측 모델을 만들거나 새로운 인사이트를 발견하는 포괄적 과정


#### 데이터 과학의 주요 구성 요소
- 데이터 수집
  - 다양한 소스에서 데이터를 모음
  - 데이터베이스, API, 웹 크롤링, 센서 등을 통해 수집
  
- 데이터 처리 및 정제
  - 결측값이나 중복값이 있는 데이터를 정제하고 필요한 형태로 변환하며 이상치를 제거
  
- 데이터 분석 및 모델링
  - 정제된 데이터를 기반으로 통계 분석을 하거나 머신러닝 알고리즘으로 데이터를 모델링하여 데이터를 이해하고 예측
  
- 데이터 시각화
  - 분석 결과를 이해하기 쉽도록 그래프 등의 시각적 도구를 활용하여 표현
  
- 의사 결정 및 인사이트 도출
  - 데이터 분석 결과를 바탕으로 비즈니스 또는 과학적 문제 해결을 위한 인사이트를 도출하고 의사 결정을 지원

#### 데이터 과학의 활용 분야
##### 마케팅 
- 고객 행동 예측, 마케팅 전략 최적화
  - 예: 고객 데이터를 분석하여 유사한 특성을 가진 고객 그룹을 식별하고, 각각의 그룹에 맞춘 마케팅 전략을 수립

##### 의료
- 질병 예측, 유전자 분석, 개인화된 치료 계획
  - 예: 환자의 유전자 정보와 의료 기록을 분석하여 특정 질병의 발병 가능성을 예측하고 맞춤형 치료 방법을 제시
  
##### 금융
- 신용 위험 분석, 금융 시장 예측, 사기 탐지
  - 예: 고객의 거래 패턴을 분석하여 비정상적인 활동을 탐지하고, 이를 통해 금융 사기를 예방
  
##### 제조 
- 예측 유지보수, 재고 관리
  - 예: 기계의 센서 데이터를 분석하여 고장 발생 가능성을 예측하고, 사전에 유지보수를 시행하여 가동 중단을 최소화
  
##### 정부 및 공공 서비스
- 범죄 예측, 교통 혼잡 예측, 환경 모니터링
  - 예: 범죄 발생 데이터를 분석하여 특정 지역에서의 범죄 발생 가능성을 예측하고, 이를 바탕으로 예방 조치를 계획
  - 예: 시간 교통 데이터와 과거 데이터를 분석하여 교통 흐름을 예측하고, 교통 체증을 줄이기 위한 최적의 경로를 제안
